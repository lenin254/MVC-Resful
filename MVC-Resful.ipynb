### Parte 2

A continuacin, creemos algunas rutas para que los usuarios puedan interactuar con la aplicacin. Aqu hay una URL de ejemplo:

```
  https://www.etsy.com/search?q=test#copy
```
Primero, especifica qu partes de la URL son  componentes segn la discusin sobre la forma de una URL. Consulta esta publicacin de [IBM](https://www.ibm.com/docs/en/cics-ts/5.1?topic=concepts-components-url) que detalla los componentes de una URL.  
**respuesta**

Protocolo (https://):
- Indica el protocolo utilizado para la comunicacin. En este caso, es HTTPS, que es una versin segura de HTTP.

Dominio (www.etsy.com):
- Es la direccin del servidor web al que se hace la solicitud.

Puerto (443):
- Es el nmero de puerto del servidor web. El puerto 443 es comnmente utilizado para HTTPS.

Ruta (/search):
- Es la parte de la URL que sigue despus del dominio y representa una ruta en el servidor. En este caso, la ruta es "/search".

Parmetros (q=test):
- Son datos adicionales que se envan al servidor, generalmente en forma de pares clave-valor. En este caso, "q" es la clave y "test" es el valor.

Fragmento (#copy):
- Representa un fragmento de la pgina web. El navegador puede desplazarse a este fragmento cuando se carga la pgina.

En Sinatra, el enrutamiento y el controlador estn acoplados, lo que facilita la declaracin de rutas. Declaramos algunas rutas RESTful para que podamos ver una lista de tareas pendientes, crear una tarea pendiente, editar una tarea pendiente y eliminar una tarea pendiente. Qu acciones RESTful deberamos utilizar para estos?

**Respuesta**  
1. **Ver Lista de Tareas Pendientes:**
   - Mtodo HTTP: GET
   - Ruta: `/tasks`
   - Accin: Mostrar todas las tareas pendientes.

2. **Crear una Nueva Tarea Pendiente:**
   - Mtodo HTTP: POST
   - Ruta: `/tasks`
   - Accin: Crear una nueva tarea pendiente.

3. **Ver Detalles de una Tarea Pendiente:**
   - Mtodo HTTP: GET
   - Ruta: `/tasks/:id`
   - Accin: Mostrar detalles de una tarea especfica por su ID.

4. **Actualizar una Tarea Pendiente:**
   - Mtodo HTTP: PUT o PATCH
   - Ruta: `/tasks/:id`
   - Accin: Actualizar una tarea especfica por su ID.

5. **Eliminar una Tarea Pendiente:**
   - Mtodo HTTP: DELETE
   - Ruta: `/tasks/:id`
   - Accin: Eliminar una tarea especfica por su ID.

### Parte 3

Dado que HTTP es un protocolo RESTful, cada solicitud debe ir seguida de una respuesta, por lo que debemos devolver una vista o redirigir a cada solicitud. Usaremos JSON para las respuestas, que es similar a lo que hacen muchas API. Hacia dnde debera ir la respuesta?.  
**Respuesta**  
La respuesta de una solicitud puede ir a:

- La pgina web actual: Para actualizar y mostrar datos en la pgina actual.
- Redireccin a otra pgina: Despus de acciones como crear o actualizar, redirigir a una pgina relevante.
- Mensajes emergentes o notificaciones: Para mostrar informacin adicional al usuario.
- Respuestas asncronas o en segundo plano: Actualizar datos en la aplicacin sin recargar la pgina.
- Archivo para descargar: Si la solicitud genera un archivo descargable.
- Respuestas a API: Para ser consumidas por otras aplicaciones o servicios (en el caso de APIs).
